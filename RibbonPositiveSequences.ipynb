```sage
Sym = SymmetricFunctions(QQ)
s = Sym.s()
e = Sym.e()
h = Sym.h()
m = Sym.m()

def ribbonBuddy(par):
    return Partition([par[i+1]-1 for i in range(len(par)-1)])

def ribbonVal(ribbon, seq):
    a = lambda i: 0 if (i<0 or i>=len(seq)) else seq[i]
    lam = ribbon[0].conjugate()
    ell = lam.length()
    mu = list(ribbon[1].conjugate()) + (ell-ribbon[1].conjugate().length())*[0]
    return matrix(ZZ, ell, ell, lambda i,j: a(lam[i]-mu[j]-i+j)).det()

def altRibbonVal(ribbon, seq):
    a = lambda i: 0 if (i<0 or i>=len(seq)) else seq[i]
    lam = ribbon[0]
    mu = ribbon[1]
    f = s(lam).skew_by(s(mu))
    return e(f).map_item(lambda b,c: (Partition([]), c*reduce(lambda x,y:x*y, [a(b[i]) for i in range(b.length())])))
    # .subs({e([i]):a(i) for i in range(lam[0]+lam.length())})
    # parent().module_morphism(on_basis = lambda b: reduce(lambda x,y:x*y, [a(b[i]) for i in range(b.length())]), codomain=ZZ)(f)
```

```sage
testSeq = [1,3,2,2,3,1]

for ribbonSize in [1,2,3,4,5,6,7,8]:
    for testRibbon in SkewPartitions(ribbonSize):
        if testRibbon.is_ribbon():
            val1 = ribbonVal(testRibbon, testSeq)
            val2 = altRibbonVal(testRibbon, testSeq)
            # if ribbonVal(testRibbon, testSeq) != altRibbonVal(testRibbon, testSeq):
            #     print(f"The functions aren't agreeing with par = {testRibbon}")
            if val1<0 or val2<0:
                print(f"At least one of the functions was negative with par = {testRibbon}")
                print(testRibbon)
                print(val1)
                print('--------------------')
```
